# Types of Plots in Matplotlib

Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Scatter Plot

```python
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.colorbar()  # Show color scale
plt.show()
```

Bar Plot

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(labels, values)
plt.show()

# Horizontal bar plot
plt.barh(labels, values)
plt.show()
```

Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, alpha=0.5, histtype='stepfilled', color='steelblue')
plt.show()
```

Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
explode = (0, 0.1, 0, 0)  # "Explode" the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```

Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```

3D Plotting with Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

ax.scatter(x, y, z, c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

#  Basic and Specialized Visualization Tools

Subplots

```python
# Create 2 subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(x, y, 'r--')
ax1.set_title('First Plot')

ax2.scatter(x, y)
ax2.set_title('Second Plot')

plt.tight_layout()
plt.show()
```

Annotations

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.annotate('local max', xy=(np.pi/2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

Customizing Styles

```python
plt.style.use('ggplot')  # Use ggplot style

x = np.random.randn(1000)
plt.hist(x, bins=30)
plt.title("Histogram with ggplot style")
plt.show()
```

# Examples

Basic Line Plot (Easy)

```python
import matplotlib.pyplot as plt
# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure and axis
fig, ax = plt.subplots()

# Plot
ax.plot(x, y, label='sin(x)', color='blue', linewidth=2)

# Customize
ax.set_title("Sine Wave")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()
ax.grid(True)

plt.show()
```

Multiple Plots with Subplots (Medium)

```python
# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create figure with 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 10))

# Plot 1
ax1.plot(x, y1, 'b-')
ax1.set_title('Sine Function')
ax1.set_ylabel('sin(x)')

# Plot 2
ax2.plot(x, y2, 'r--')
ax2.set_title('Cosine Function')
ax2.set_ylabel('cos(x)')

# Plot 3
ax3.plot(x, y3, 'g:')
ax3.set_title('Tangent Function')
ax3.set_ylabel('tan(x)')
ax3.set_xlabel('x')

# Adjust layout
plt.tight_layout()
plt.show()
```

Advanced Customization (Medium-Hard)

```python
# Generate data
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1.5, 1000)
data3 = np.random.normal(-2, 0.5, 1000)

# Create figure with custom size
fig = plt.figure(figsize=(12, 8))

# Create gridspec for complex layout
gs = fig.add_gridspec(2, 2, width_ratios=[3, 1], height_ratios=[1, 3])

# Main plot
ax1 = fig.add_subplot(gs[1, 0])
ax1.scatter(data1, data2, alpha=0.6, c='blue', edgecolor='black')
ax1.set_xlabel('Data 1')
ax1.set_ylabel('Data 2')
ax1.set_title('Scatter Plot with Marginal Distributions')

# Top histogram (data1)
ax2 = fig.add_subplot(gs[0, 0], sharex=ax1)
ax2.hist(data1, bins=30, color='blue', alpha=0.7)
ax2.set_ylabel('Count')
ax2.set_title('Distribution of Data 1')

# Right histogram (data2)
ax3 = fig.add_subplot(gs[1, 1], sharey=ax1)
ax3.hist(data2, bins=30, orientation='horizontal', color='green', alpha=0.7)
ax3.set_xlabel('Count')

# Add a boxplot
ax4 = fig.add_subplot(gs[0, 1])
ax4.boxplot([data1, data2, data3], labels=['Data 1', 'Data 2', 'Data 3'])
ax4.set_title('Boxplot Comparison')

# Adjust layout
plt.tight_layout()
plt.show()
```

Interactive Plot with Widgets (Hard)

```python
from ipywidgets import interact

def plot_function(a=1, b=0, c=0):
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'{a}x + {b}x + {c}', linewidth=2)
    plt.title('Quadratic Function')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.ylim(-10, 10)
    plt.show()

interact(plot_function, a=(-2, 2, 0.1), b=(-5, 5, 0.1), c=(-5, 5, 0.1))
```

Advanced 3D Visualization (Hard)

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Create figure
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Create data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot surface
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize z axis
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=5)

# Add contour projections
offset = -1.5
ax.contour(X, Y, Z, zdir='z', offset=offset, cmap=cm.coolwarm)
ax.contour(X, Y, Z, zdir='x', offset=-6, cmap=cm.coolwarm)
ax.contour(X, Y, Z, zdir='y', offset=6, cmap=cm.coolwarm)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface with Contour Projections')
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_zlim(offset, 1.5)

plt.show()
```
